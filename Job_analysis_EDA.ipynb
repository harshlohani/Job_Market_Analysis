```python
def replace_substring_if_exists(data1, location, substring_to_find, replacement_substring):
    
    data2 = data1.copy()  # Create a copy to avoid modifying the original DataFrame
    data2['location'] = data1['location'].str.replace(substring_to_find, replacement_substring)
    return data2
```

```python
print(data2['location'].unique())
```

```python
data3=data2.copy()
data3['location']=data2['location'].str.replace(r'[-\s/]', '')
data3['location']=data2['location'].str.replace("''", "")
data3['location']=data3['location'].str.replace('-', '')
data3['location']=data3['location'].str.replace('(USA)', '')
data3['location']=data3['location'].str.replace('(Kondapur)', '')
data3['location']=data3['location'].str.replace('(All Areas)', '')
data3['location'] = data3['location'].str.replace(r'\(.*?\)', '', regex=True)
```

```python
print(data3['location'])
```

```python
unique_substrings = set(','.join(data3['location']).split(','))
```

```python
unique_substrings=pd.DataFrame(unique_substrings)
unique_substrings
```

```python
unique_substrings.columns = ['Cities']
```
